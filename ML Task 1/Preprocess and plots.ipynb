```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the wine dataset
wine_data = pd.read_csv('winequality-red.csv')

# Display the first few rows to understand the data
print("First 5 rows of the dataset:")
display(wine_data.head())

# Check for NaN and infinite values
print("\nNaN values in each column:\n", wine_data.isna().sum())
print("\nInfinite values in each column:\n", np.isinf(wine_data).sum())

# Handle NaN and infinite values
wine_data.replace([np.inf, -np.inf], np.nan, inplace=True)
wine_data.dropna(inplace=True)

# Check data types
print("\nData types:\n", wine_data.dtypes)
```

```python
# 1. Correlation Matrix Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = wine_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 2. Pairplot of Features
g = sns.pairplot(wine_data, diag_kind='kde', height=2.5)
g.fig.suptitle("Pairplot of Features", y=1.02)  # Adjust y for title spacing
plt.show()
```

```python
# 3. Scatter Plots: Each Feature vs. Target (Quality)
for column in wine_data.columns:
    if column != 'quality':  # Skip target column for x-axis
        plt.figure(figsize=(6, 4))
        sns.scatterplot(data=wine_data, x=column, y='quality', alpha=0.5)
        plt.title(f'{column} vs. Quality')
        plt.show()
```

```python
# 4. Boxplots: Feature Distributions by Quality Level
for column in wine_data.columns:
    if column != 'quality':  # Skip target column
        plt.figure(figsize=(6, 4))
        sns.boxplot(x='quality', y=column, data=wine_data)
        plt.title(f'Boxplot of {column} by Quality')
        plt.show()
```
